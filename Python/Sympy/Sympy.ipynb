```python
import sympy
```

```python
from sympy.solvers import solve
from sympy import Symbol
from sympy import Eq
```

```python
x = Symbol('x')
eq = (x**2 - 1, x)
solve(eq)
```

```python
y = Symbol('y')
eq = Eq(y + 3 + 8)
solve(eq)
```

```python
solve(x < 3)
```

```python
solve([x**2 - 3, y - 1], set=True)
```
